```python
import tkinter as tk
from tkinter import messagebox
import csv
import os

CSV_FILE = "ab_testing_results.csv"

if not os.path.exists(CSV_FILE):
    with open(CSV_FILE, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Layout", "Vote"])

while True:
    root = tk.Tk()
    root.title("A/B Testing - Choose Layout")

    tk.Label(root, text="UI Layout A", font=("Arial", 14)).grid(row=0, column=0, padx=20, pady=10)
    tk.Label(root, text="UI Layout B", font=("Arial", 14)).grid(row=0, column=1, padx=20, pady=10)

    tk.Label(root, text="Layout A: Simple, Minimalistic", bg="lightblue", width=25, height=5).grid(row=1, column=0, padx=20, pady=10)
    tk.Label(root, text="Layout B: Colorful, Modern", bg="lightgreen", width=25, height=5).grid(row=1, column=1, padx=20, pady=10)

    def select_layout(layout):
        with open(CSV_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([layout, 1])
        root.destroy()

        preferences = {"Layout A": 0, "Layout B": 0}
        with open(CSV_FILE, mode="r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if row["Layout"] in preferences:
                    preferences[row["Layout"]] += 1

        summary = f"Layout A: {preferences['Layout A']} votes\nLayout B: {preferences['Layout B']} votes\n"
        if preferences["Layout A"] > preferences["Layout B"]:
            summary += "\nLayout A is more preferred!"
        elif preferences["Layout A"] < preferences["Layout B"]:
            summary += "\nLayout B is more preferred!"
        else:
            summary += "\nBoth layouts are equally preferred!"

        messagebox.showinfo("Summary Report", summary)

    tk.Button(root, text="Select Layout A", command=lambda: select_layout("Layout A")).grid(row=2, column=0, pady=20)
    tk.Button(root, text="Select Layout B", command=lambda: select_layout("Layout B")).grid(row=2, column=1, pady=20)

    root.mainloop()

    again = messagebox.askyesno("Continue?", "Do you want to vote again?")
    if not again:
        break
```
