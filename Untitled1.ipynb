```python
import tkinter as tk
from tkinter import messagebox as mb
import csv, os

csv1 = "ab_testing.csv"
if not os.path.exists(csv1):
    with open(csv1, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Layout", "Vote"])

def select_layout(layout):
    # Record vote
    with open(csv1, "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([layout, 1])

    # Count votes
    p = {"Layout A": 0, "Layout B": 0}
    with open(csv1, "r", newline="") as f:
        reader = csv.DictReader(f)
        for r in reader:
            if r["Layout"] in p:
                p[r["Layout"]] += 1

    # Show summary
    summary = f"Layout A: {p['Layout A']} votes\nLayout B: {p['Layout B']} votes\n"
    if p["Layout A"] > p["Layout B"]:
        summary += "Layout A is more preferred"
    elif p["Layout A"] < p["Layout B"]:
        summary += "Layout B is more preferred"
    else:
        summary += "Both layouts are equally preferred"

    mb.showinfo("Summary", summary)

    # Ask to continue
    again = mb.askyesno("Continue?", "Vote again?")
    if not again:
        root.destroy()

# Create one window only
root = tk.Tk()
root.geometry("600x700")
root.title("A/B Testing Vote")

tk.Label(root, text="Layout A", font=("Arial", 14)).grid(row=0, column=0, padx=20, pady=10)
tk.Label(root, text="Layout B", font=("Arial", 14)).grid(row=0, column=1, padx=20, pady=10)
tk.Label(root, text="Layout A: Minimalistic design", font=("Arial", 14), bg="yellow", width=45, height=5).grid(row=1, column=0, padx=20, pady=10)
tk.Label(root, text="Layout B: Bright Layout", font=("Arial", 14), bg="red", width=45, height=5).grid(row=1, column=1, padx=20, pady=10)

tk.Button(root, text="Vote Layout A", command=lambda: select_layout("Layout A")).grid(row=2, column=0, padx=20, pady=10)
tk.Button(root, text="Vote Layout B", command=lambda: select_layout("Layout B")).grid(row=2, column=1, padx=20, pady=10)

root.mainloop()
```
