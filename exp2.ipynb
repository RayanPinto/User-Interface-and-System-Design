```python
chat_data=[
    {
        "q":"what is python",
        "s":"llllll",
        "l":"123456",
    }
    ,
    {"q":"what is java",
        "s":"2222222",
        "l":"1234567890",
    },{
        "q":"what is go",
        "s":"llllll11111111111111111",
        "l":"123456111111111111111",
    }
]
sc=0
lc=0
for h in chat_data:
    print(h["q"])
    print(h["s"])
    print(h["l"])
    while True:
        v=int(input("Enter a value"))
        if v==1:
            sc=sc+1
            break
        else:
            lc+=1
            break
if sc>lc:
    print("short is preferred",sc,lc)
else:
    print("long is prefeered",sc,lc)
```
