```python
import tkinter as tk
from textblob import TextBlob
import re
def theme(ss):
    if ss>0.2:
        root.configure(bg="yellow")
        res_lab.config(bg="red",fg="blue",font=("Arial",14))
        return "Positive Mood: Bright Colors and colorful layout"
    elif ss<-0.2:
        root.configure(bg="red")
        res_lab.config(bg="yellow",fg="blue",font=("Arial",14))
        return "Negative Mood: dull Colors and darker layout"
    else:
        root.configure(bg="grey")
        res_lab.config(bg="yellow",fg="blue",font=("Arial",14))
        return "Neutral Mood: normal Colors and normal layout"
def ismeaning(text):
    w=re.findall(r'[a-zA-Z]+',text)
    return len(w)>0
def analyze():
    sen=entry.get().strip()
    if not sen:
        res_lab.config(text="hii")
        return
    if not ismeaning(sen):
        res_lab.config(text="bye")
        return
    anal=TextBlob(sen)
    ss=anal.sentiment.polarity
    if ss==0.0 and len(sen.split())<=1:
        res_lab.config(text="rayan")
        return
    
    th=theme(ss)
    res_lab.config(text=f"sentiment score:{ss:.2f}\n{th}")
    return
root=tk.Tk()
root.title("User sentiment")
root.geometry("600x700")
root.configure(bg="brown")
title=tk.Label(root,text="Enter",bg="blue",font=("Arial",14))
title.pack(pady=10)
entry=tk.Entry(root,text="Enter",font=("Arial",14))
entry.pack(pady=10)
button=tk.Button(root,text="anal",font=("Arial",14),command=analyze)
button.pack(pady=10)
res_lab=tk.Label(root,text="",bg="blue",font=("Arial",14))
res_lab.pack(pady=10)
root.mainloop()
        
```
